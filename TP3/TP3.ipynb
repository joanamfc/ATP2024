## TPC3: Jogo dos 21 fósforos

Desenvolve em Python o código necessário para o jogo dos 21 fósforos: 

### O jogo

* No início do jogo, há 21 fósforos;
* Cada jogador (computador ou utilizador), pode tirar 1, 2, 3 ou 4 fósforos quando for a sua vez de jogar;
* Os jogadores jogam alternadamente;
* **Quem tirar o último fósforo perde!**

### O programa 

* O jogo deverá ter dois modos: o jogador joga em primeiro lugar e o computador começa a jogar em segundo lugar e, no segundo modo, o computador começa em primeiro; 
* Quando o computador começa a jogar em segundo lugar, deve ganhar sempre o jogo;
* Quando o computador começa a jogar em primeiro lugar, se o utilizador cometer um erro de cálculo, o computador deverá passar para a posição de vencedor e ganhar o jogo.

```python

from random import randint
fósforos = 21
nome = input("Coloque o seu nome de usuário:")
print(f"Bem-vindo {nome}, vamos jogar o jogo dos 21 fósforos! Começamos com 21 fósforos, cada jogador deve retirar de 1 a 4 fósforos, quem retirar o último perde.")
print("Escolhe uma modalidade: 1. Tu começas a jogar em primeiro. 2. O computador joga em primeiro lugar.")
escolha = input("Escolhe a modalidade 1 ou a modalidade 2: ")

if escolha == "1" :
    while fósforos > 1:
        retirados = int(input("Introduça o número de fósforos que quer retirar, pode retirar de 1 a 4 fósforos."))
        while retirados not in [1,2,3,4]:
            print("Número de fósforos invalido, volte a introduzir uma quantidade e 1 a 4.")
            retirados = int(input("Introduça o número de fósforos que quer retirar, pode retirar de 1 a 4 fósforos."))
        fósforos = fósforos - retirados
        print(f"Retiraste {retirados}, sobraram {fósforos} fósforos.")
        fósforos = fósforos - (5- retirados)
        print(f"Eu retirei {5 - retirados}, sobraram então {fósforos} fósforos.")
    retirados = int(input("Introduça o número de fósforos que quer retirar, pode retirar de 1 a 4 fósforos."))
    while retirados > fósforos:
        print("Ora bem, já sabes o que vai acontecer...")
        retirados = int(input("Introduça o número de fósforos que quer retirar, pode retirar de 1 a 4 fósforos."))
    print(f"Retiraste {retirados}, não sobrou nenhum fósforo, logo o vencedor fui eu! Mais sorte para a próxima...")

elif escolha == "2":
    print("Sou eu a começar")
    retirados_computador = randint(1,4)
    fósforos = 21 - retirados_computador
    print(f"Retirei { retirados_computador } fósforos, sobraram { fósforos } fósforos.")
    while fósforos > 0:
        usuário = int(input("E tu, quantos fósforos queres retirar?"))
        if usuário in [1,2,3,4]:
            fósforos = fósforos - usuário
            print(f"Tu retiraste {usuário} fósforos, ainda faltam {fósforos} fósforos.")
            if fósforos % 5 == 0 and fósforos >= 6:
                retirados_computador = randint(1,4)
                fósforos =fósforos - retirados_computador
                print(f"Eu vou retirar {retirados_computador}, por isso sobram {fósforos}.")
            elif fósforos % 5 != 0 and fósforos >= 6:
                computador = (fósforos % 5)
                fósforos = fósforos - computador
                print(f"Eu vou retirar {computador} e ainda sobram {fósforos}.")
            elif fósforos < 6 and fósforos > 1:
                print(f"Eu vou retirar {fósforos - 1}, ainda falta 1 fósforo.")
                fósforos = 1
                usuário = int(input("E tu, quantos fósforos queres retirar?"))
                if usuário == 1:
                    print("Ora bem, não sobrou nenhum fósforo, tu retiraste o último... Eu Ganhei! Mais sorte para a próxima...")
                    fósforos = 0
                else:
                    while usuário != 1:
                        usuário = int(input("Agora é a tua vez, quantos fósforos vais tirar?"))
                        if usuário == 1:
                            print("Ora bem, não sobrou nenhum fósforo, tu retiraste o último... Eu Ganhei! Mais sorte para a próxima...")
                            fósforos = 0
            elif fósforos == 1:
                print("Ora bem, falta um fósforo...")
                print("Eu retirei o último, os fósforos acabaram... Parabéns, tu ganhaste!")
                fósforos = 0
else:
    print("Escolha um modo válido...")

    
```
